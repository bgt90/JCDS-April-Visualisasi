## Histogram

### Using Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#load dataset

tips= sns.load_dataset('tips')
tips.head()
```

```python
# histogram dari total_bill

plt.hist(tips['total_bill'])
plt.title('Total Bill Distribution',size=20) #judul tabel
plt.xlabel('Total Bill')# label x axis
plt.ylabel('Frequency')# label y axis
plt.show() #langsung memunculkan grafik, tanpa menampilkan background process
```

```python
plt.style.use('seaborn') #style dari tampilan grafik
plt.figure(figsize=(8,6)) #ukuran dari grafik
plt.hist(tips['total_bill'],20, color= 'green') #data, jumlah bins, color= warna
plt.title('Total Bill Distribution',size=20)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.grid(False) #grid di background
plt.legend(['Total Bill for Customer'], loc=0) #memunculkan legend, lokasi
# plt.savefig('distribusi.png') #untuk menyimpan gambar, lakukan sebelum plt.show
plt.show()
```

## Line Plot

### Using Matplotlib

```python
flights[flights['month']=='January']['year']
```

```python
flights[flights['month']=='January']['passengers']
```

```python
x1=flights[flights['month']=='January']['year']
y1=flights[flights['month']=='January']['passengers']
x2=flights[flights['month']=='December']['year']
y2=flights[flights['month']=='December']['passengers']

plt.figure(figsize=(12,6))
plt.plot(x1,y1, 'r',
        x2,y2, 'b--')
plt.title('Annual Number of passangers in January and December', size=20)
plt.xlabel('year')
plt.ylabel('passangers')
plt.legend(['January','December'])
plt.show()
```

## Piechart

### Using Matplotlib

```python
tips.head()
```

```python
day_proportion=pd.DataFrame(tips['day'].value_counts()).reset_index()
day_proportion
```

```python
plt.figure(figsize=(8,6))
plt.pie(
    day_proportion['day'],
    labels=day_proportion['index'],
    autopct='%.2f', #memberi label data, 2 angka di belakang koma
    explode=(0,0,0,0.2) # memberi jarak antar slice, digunakan untuk menonjolkan slice tertentu
)

plt.axis('equal')
plt.title('Piechart: Consumers Proportions by Day', size=15)
plt.show()
```

### Using Pandas

```python
day_proportion=pd.DataFrame(tips['day'].value_counts())
day_proportion
```

```python
def absolute_value(val):
    a  = np.round(val/100.*day_proportion['day'].sum(), 0)
    return a
```

```python
absolute_value(day_proportion['day'])
```

```python
day_proportion.plot.pie(
    y='day',
    figsize=(10,10),
    autopct= absolute_value,
    explode=(0.2,0,0,0)
)
plt.show()
```

### Choropeth

Data Source:  
* Jakarta Open Data, https://data.jakarta.go.id/dataset/pelaporan-perpindahan-penduduk-keluar-provinsi-dki-jakarta-tahun-2020-berdasarkan-kelurahan/resource/134253ff-2915-44b8-a1b3-2b1869366ed0  
* Jakarta by Kelurahan geo public, https://pstyd.carto.com/tables/jakarta_by_kelurahan_geo/public 


```python
df=pd.read_csv('pendatang_DKI_Maret_2020.csv')
df.head()
```

```python
df_kel=df[['kelurahan','jumlah']]
df_kel.head()
```

```python
(df.kelurahan=='SETIA BUDI').sum()
```

```python
import json
jason=json.load(open('jakarta_by_kelurahan_geo.geojson'))
```

```python
jason['features'][0]['properties']['name']
```

```python
nama=[]
for i in range(len(jason['features'])):
    nama.append(jason['features'][i]['properties']['name'])
```

```python
(df_kel['kelurahan']=='PULO GADUNG').sum()
```

```python
nama.index('PULOGADUNG')# cara penulisan di data geojson dan df penduduk berbeda
```

```python
# jakarta geojson
jakarta=r'jakarta_by_kelurahan_geo.geojson'
jakarta_location=[-6.2088, 106.8456]

jakarta_map=folium.Map(location=jakarta_location, zoom_start=12)

folium.Choropleth(geo_data=jakarta,
                 data= df_kel,
                 columns=['kelurahan','jumlah'],
                 key_on='feature.properties.name',
                 fill_color='YlOrRd',
                  fill_opacity=0.7,
                  line_opacity=0.2,
                  legend_name='Jumlah Pendatang di Jakarta'
                 ).add_to(jakarta_map)

jakarta_map
```

```python
bins = list(df_kel['jumlah'].quantile([0, 0.25, 0.5, 0.75, 1]))
jakarta=r'jakarta_by_kelurahan_geo.geojson'
jakarta_location=[-6.2088, 106.8456]

jakarta_map=folium.Map(location=jakarta_location, zoom_start=12)

folium.Choropleth(geo_data=jakarta,
                 data= df_kel,
                 columns=['kelurahan','jumlah'],
                 key_on='feature.properties.name',
                 fill_color='BuPu',
                  fill_opacity=0.7,
                  line_opacity=0.2,
                  legend_name='Jumlah Pendatang di Jakarta',
                  bins=bins
                 ).add_to(jakarta_map)

jakarta_map
```
